```python
def akurasi(mat) :
    p,l=mat.shape
    data=0
    for i in range (p):
        data=data+mat[i,i]
    return data/sum(mat.flatten())

def getkol(mat,j):
    p,l= mat.shape
    data=0
    for i in range(p):
        if(i!=j):
            data=data+mat[i,j]
    return data

def getbar(mat,j):
    p,l= mat.shape
    data=0
    for i in range(p):
        if(i!=j):
            data=data+mat[j,i]
    return data


def presisi (mat):
    p,l=mat.shape
    data=0
    pres = []
    for i in range (p):
        data=mat[i,i]/(mat[i,i]+getkol(mat,i))
        pres.append(data)
    return (sum(pres)/p)*100

def recall (mat):
    p,l=mat.shape
    data=0
    pres = []
    for i in range (p):
        data=mat[i,i]/(mat[i,i]+getbar(mat,i))
        pres.append(data)
    return (sum(pres)/p)*100
```
